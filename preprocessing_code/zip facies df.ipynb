```python
import lzma 
```

```python

# open input file as binary and read input data
with open('mcmurray_facies_v1.csv', 'rb') as iFile:
    iData = iFile.read()

# compress data
oData = lzma.compress(iData)

# open output file as binary and write compressed data
with open('mcmurray_facies_v1.zip', 'wb') as oFile:
    oFile.write(oData)
```
